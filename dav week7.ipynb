```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Polygon

# COVID-19 case data
data = {
    "State": ["Maharashtra", "Kerala", "Delhi", "Karnataka", "Tamil Nadu"],
    "Cases": [320, 250, 150, 180, 200]
}
df = pd.DataFrame(data)

# Dummy geometries for demonstration (not accurate)
geometry = [
    Polygon([(72, 18), (73, 18), (73, 19), (72, 19)]),  # Maharashtra
    Polygon([(76, 10), (77, 10), (77, 11), (76, 11)]),  # Kerala
    Polygon([(77, 28), (78, 28), (78, 29), (77, 29)]),  # Delhi
    Polygon([(75, 12), (76, 12), (76, 13), (75, 13)]),  # Karnataka
    Polygon([(78, 11), (79, 11), (79, 12), (78, 12)])   # Tamil Nadu
]

# Create GeoDataFrame
geo_df = gpd.GeoDataFrame(df, geometry=geometry)

# Plotting
plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots(figsize=(10, 12))
geo_df.plot(column='Cases', cmap='Reds', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

# Annotate state names
for idx, row in geo_df.iterrows():
    centroid = row['geometry'].centroid
    ax.text(centroid.x, centroid.y, row['State'], fontsize=10, ha='center', va='center', color='black')

# Title and layout
ax.set_title("COVID-19 Cases Across Indian States (2020)", fontsize=16)
ax.axis('off')

# ShowÂ plot
plt.show()
```
